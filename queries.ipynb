## 2. More advanced queries (including joins)
* get the 10 most expensive neighborhoods
* get the number of neighbourhoods
* get all the information from the airbnb table about listings from persons called Jennifer

```python
query = "SELECT avg(price), neighbourhood FROM airbnb GROUP BY neighbourhood  ORDER BY avg(price) DESC LIMIT 10"

cursor.execute(query)

records = pd.DataFrame(cursor.fetchall())

records.head(1000)
```

```python
query = "SELECT COUNT(DISTINCT neighbourhood) FROM airbnb;"

cursor.execute(query)

records = pd.DataFrame(cursor.fetchall())

records.head(1000)
```

```python
query = "SELECT * FROM airbnb \
        INNER JOIN airbnb_host ON airbnb.id = airbnb_host.id \
        WHERE airbnb_host.host_name IN ('Jennifer');"

cursor.execute(query)

records = pd.DataFrame(cursor.fetchall())

records.head()
```
